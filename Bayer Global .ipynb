### Key Peforming Metrics

#### Sales Revenue

First step in our calculations is to total sales made and building up from there

```python
#calculating the total sales made for each year
merged_crm_sap_df["Sales made 2020"] = merged_crm_sap_df["volume sales 2020 (units)"] * merged_crm_sap_df["Price 2020 (units)"]
merged_crm_sap_df["Sales made 2021"] = merged_crm_sap_df["volume sales 2021(units)"] * merged_crm_sap_df["Price 2021 (units)"]
merged_crm_sap_df["Sales made 2022"] = merged_crm_sap_df["volume sales 2022 (units)"] * merged_crm_sap_df["Price 2022 (units)"]
```

```python
#check the head of merged_crm_sap_df
merged_crm_sap_df.head()
```

Since G2N and P&I are both expenses from our brand, we subtract them from the sales made for each year

```python
#Revenue from sales after G2N 
merged_crm_sap_df["G2N 2020"] = merged_crm_sap_df["Sales made 2020"] * merged_crm_sap_df["G2N"]
merged_crm_sap_df["G2N 2021"] = merged_crm_sap_df["Sales made 2021"] * merged_crm_sap_df["G2N"]
merged_crm_sap_df["G2N 2022"] = merged_crm_sap_df["Sales made 2022"] * merged_crm_sap_df["G2N"]
```

```python
#check head of data
merged_crm_sap_df.head()
```

```python
#Total revenue from sales for each year
merged_crm_sap_df["Total revenue sales 2020"] = merged_crm_sap_df["Sales made 2020"] - (merged_crm_sap_df["G2N 2020"] + merged_crm_sap_df["P&I"])
merged_crm_sap_df["Total revenue sales 2021"] = merged_crm_sap_df["Sales made 2021"] - (merged_crm_sap_df["G2N 2021"] + merged_crm_sap_df["P&I"])
merged_crm_sap_df["Total revenue sales 2022"] = merged_crm_sap_df["Sales made 2022"] - (merged_crm_sap_df["G2N 2022"] + merged_crm_sap_df["P&I"])
```

```python
#checking the head of the data
merged_crm_sap_df.head()
```

Now we can calculate the total revenue gotten from sales in each year

```python
#calculating the revenue for 2020
revenue_2020= merged_crm_sap_df["Total revenue sales 2020"].sum().round(2)
print(f"The Revenue generated in 2020 through sales is ${revenue_2020}")
```

```python
#calculating the revenue for 2021
revenue_2021= merged_crm_sap_df["Total revenue sales 2021"].sum().round(2)
print(f"The Revenue generated in 2021 through sales is ${revenue_2021}")
```

```python
#calculating the revenue for 2022
revenue_2022= merged_crm_sap_df["Total revenue sales 2022"].sum().round(2)
print(f"The Revenue generated in 2022 through sales is ${revenue_2022}")
```

From the calculations above, we can see that there is a gradual decrease in the revenue generated from sales. This can be tied to the huge reudction in purchases from our high risk and vulnerable customers and our reluctant customers not improving their purchases

Also with this calculations above we can measure the first Key performing metric for the sales team which is Sales revenue and comparing the values we can see that the sales revenue decrease over the years from 2020 to 2022.

```python
sales_percent_21_22 = (((revenue_2022  - revenue_2021)/ revenue_2021)*100).round(1)
print(f"The Percentage growth in sales from 2021 to 2022 is {sales_percent_21_22}%")
```
